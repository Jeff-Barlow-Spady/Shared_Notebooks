```python
## Plotnine Basic Graphing Demo

from plotnine import ggplot, geom_bar, geom_point,geom_line, geom_histogram,\
geom_boxplot, facet_wrap, theme_minimal, data, aes, theme_seaborn, geom_smooth 

"""
ggplot: Creates a new plotnine plot

geom_bar: Adds a bar chart layer to the plot 
- Takes a categorical variable mapped to x to create the bars
- Can specify stat to compute counts (stat='count') or proportions (stat='identity')

geom_point: Adds a scatter plot layer to the plot
- Takes x and y variables to determine the location of points

geom_line: Adds a line plot layer to the plot
- Takes x and y variables to determine the line path

geom_histogram: Adds a histogram layer 
- Takes a continuous variable mapped to x to bin into histogram  
- Can specify bins for number of bins

geom_boxplot: Adds a boxplot layer
- Takes a categorical variable mapped to x and a continuous variable mapped to y

facet_wrap: Wraps a grid of panels onto the plot
- Takes a categorical variable to split data across panels

theme_minimal: Applies a minimal plotting theme to the plot

data: Specifies the dataframe to use for the plot layers

aes: Specifies the aesthetic mappings of variables to plotting attributes

theme_seaborn: Applies the seaborn plotting style 

geom_smooth: Adds a smoothed conditional mean line
- Takes x and y variables to determine the line path
"""
```

```python
#Create dataframes with sample data
df = data.diamonds
df2 = data.faithful
```

```python
### Scatter Plot
ggplot(df, mapping=aes(x='price', y='carat', color='cut')) + geom_point(color='blue') + theme_seaborn(style="darkgrid",) + geom_line()

"""
The first line of code loads the df data frame into the ggplot() function.

The mapping argument specifies the aesthetic mappings for the plot. In this case, the x-axis is mapped to the price column, 
the y-axis is mapped to the carat column, and the color of the points is mapped to the cut column.

The geom_point() function adds a layer of points to the plot, 
and the color argument specifies that the points should be colored blue.

The theme_seaborn() function sets the theme of the plot to the seaborn darkgrid style.

Finally, the geom_line() function adds a line to the plot, connecting the points.

"""
```

```python
### Line Plot

ggplot(df, aes(x='x', y='y', color='price')) + geom_line(color='blue') + theme_minimal()

""""
The first line of code loads the df data frame into the ggplot() function.

The aes() function specifies the aesthetic mappings for the plot. In this case, the x-axis is mapped to the x column, the y-axis is mapped to the y column, and the color of the line is mapped to the price column.

The geom_line() function adds a layer of lines to the plot, and the color argument specifies that the lines should be colored blue.

The theme_minimal() function sets the theme of the plot to the minimal style.

"""
```

```python

### Bar Plot

ggplot(df2, aes(x='waiting', y='eruptions', color=max(df2['eruptions']))) + geom_bar(stat='identity') + theme_minimal()

"""""
The first line of code loads the df2 data frame into the ggplot() function.

The aes() function specifies the aesthetic mappings for the plot. In this case, the x-axis is mapped to the waiting column, the y-axis is mapped to the eruptions column, and the color of the bars is mapped to the maximum value of the eruptions column.

The geom_bar() function adds a layer of bars to the plot, and the stat='identity' argument specifies that the bars should be drawn as is, without any transformations.

The theme_minimal() function sets the theme of the plot to the minimal style.
"""
```

```python
### Histogram

from plotnine import theme_538, theme_matplotlib
"""
These themes can be used to customize the appearance of plots created using the plotnine library.

The theme_538 theme is based on the visual style of the popular data journalism website FiveThirtyEight, while the theme_matplotlib theme is based on the visual style of the popular Python plotting library Matplotlib.

To use these themes, you can simply pass them as an argument to the theme() function when creating a plot.
"""
```

```python
ggplot(df,aes(x='price')) + geom_histogram(binwidth=50) + theme_matplotlib()
""""
Creates a histogram of the price column in a DataFrame called df using the popular Python library ggplot.

The aes() function is used to specify the aesthetic mappings for the plot.

In this case, the x-axis is mapped to the price column, and the y-axis is mapped to the count of prices.

The geom_histogram() function is used to create the histogram, and the binwidth argument is used to specify the width of the bins.

The theme_matplotlib() function is used to apply the matplotlib theme to the plot.

This theme changes the appearance of the plot to match the style of matplotlib plots.

"""
```

```python
### Box Plot

ggplot(df, aes(x='color', y='price')) + geom_boxplot() + theme_minimal()
"""
The aes() function is used to specify the aesthetic mappings for the plot.

The geom_boxplot() function creates...a box plot haha

The theme_minimal() function sets the theme of the plot to the minimal style.
"""
```

```python
### Faceted Plot

ggplot(df, aes(x='cut', y='price',color='color')) + geom_point() + facet_wrap('color') + theme_minimal()
""""
This code creates a scatter plot of the price column in a DataFrame called df, grouped by the values in the color column.

The aes() function is used to specify the aesthetic mappings for the plot.

In this case, the x-axis is mapped to the cut column, the y-axis is mapped to the price column, and the color of the points is mapped to the color column.

The geom_point() function is used to create the scatter plot.

The facet_wrap() function is used to wrap the plot by the values in the color column, creating a separate plot for each value.

The theme_minimal() function is used to apply a minimalist theme to the plot.
"""
```

```python

#lol

def plot_3d_scatterplot(x_data, y_data, z_data, title="", x_label="", y_label="", z_label="", marker_color='deeppink', **plot_kwargs):
    import plotly.graph_objs as go
    import plotly.express as px
    import matplotlib as plt
    """Plots a 3D scatter plot.

    Args:
        x_data (list): The x-axis data.
        y_data (list): The y-axis data.
        z_data (list): The z-axis data.
        title (str): The title of the chart.
        x_label (str): The label for the x-axis. 
        y_label (str): The label for the y-axis.
        z_label (str): The label for the z-axis.
        marker_color (str): Defaults to deeppink.
        plot_kwargs 
    
    """
    fig = go.Figure(data=[go.Scatter3d(
        x=x_data, 
        y=y_data, 
        z=z_data,
        mode='markers',
        marker_color=marker_color
    )])
    fig.update_layout(
        scene=dict(
            xaxis_title=x_label,
            yaxis_title=y_label,
            zaxis_title=z_label
        ),
        title=title
    )
    return fig

""""
The function plot_3d_scatterplot takes in data for the x, y, and z axes, as well as a title, axis labels, and a marker color.

It then uses the plotly.graph_objs library to create a 3D scatter plot.

The function first creates a plotly.graph_objs.Scatter3d object, which is used to represent the scatter plot.

The object is then configured with the data for the x, y, and z axes, as well as the marker color.

The function then creates a plotly.graph_objs.Layout object, which is used to configure the layout of the plot.

The layout object is configured with the axis labels and the title.

Finally, the function uses the plotly.offline.plot function to display the plot.
"""
```
